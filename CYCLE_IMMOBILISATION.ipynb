```python
import pandasql as psql
C = '''
SELECT *
FROM merge
WHERE ("Compte Général - Code" BETWEEN 211300010 AND 283560214)
   OR ("Compte Général - Code" BETWEEN 619120000 AND 619600000)

'''


D = psql.sqldf(C, locals())

NEW  =D['Solde_Compte_2024']-D['Solde_Compte_2023']

```

```python
D['DEFFERENCE'] = NEW
D
```

```python
PERCENTAGE = (D['DEFFERENCE']/D['Solde_Compte_2023'])*100
D['PERCENTAGE'] = PERCENTAGE
D
```

```python
D['PERCENTAGE'] = D['PERCENTAGE'].apply(lambda x: f'{x:.2f}%')
```

```python

ACQUISITION_IMMO = pd.read_excel('Cycle immobilisations 07-2024.xlsx', sheet_name='ACQUISITION IMMO')
ACQUISITION_IMMO
```

```python

CESSION_IMMO = pd.read_excel('Cycle immobilisations 07-2024.xlsx', sheet_name='CESSION IMMO')
CESSION_IMMO
```

```python
M = '''
SELECT "Compte",
       "Date d'acqu.",
       "Montant d'acqu."
FROM ACQUISITION_IMMO
WHERE ("Compte" BETWEEN 211300010 AND 283560214
       OR "Compte" BETWEEN 619120000 AND 619600000)


GROUP BY Compte
'''


N = psql.sqldf(M, locals())
N
```

```python
V = '''
SELECT "Compte",
       "Date d'acqu.",
       "Date Cession",
       "Produit de Cession"
FROM CESSION_IMMO
WHERE ("Compte" BETWEEN 211300010 AND 283560214
       OR "Compte" BETWEEN 619120000 AND 619600000)



'''


L = psql.sqldf(V, locals())
L
```
